```python
import math
from fractions import Fraction

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b != 0:
        return a / b
    else:
        return "Cannot divide by zero!"

def cast_input(prompt):
    return float(input(prompt))

def cast_input_on_same_line(prompt):
    return map(float, input(prompt).split())

def orders_of_operations(expression):
    return eval(expression)

def modulus_and_remainder(a, b):
    return a % b, a // b

def modulus_and_factors(n):
    factors = []
    for i in range(1, int(math.sqrt(n)) + 1):
        if n % i == 0:
            factors.append(i)
            factors.append(n // i)
    return factors

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def find_factors(n):
    factors = []
    for i in range(1, n + 1):
        if n % i == 0:
            factors.append(i)
    return factors

def reciprocals(a, b):
    if a != 0:
        return 1/a, 1/b
    else:
        return "Cannot find reciprocal of zero!"

def square(n):
    return n ** 2

def square_root(n):
    return math.sqrt(n)

def floor(n):
    return math.floor(n)

def square_factors(n):
    factors = []
    for i in range(1, int(math.sqrt(n)) + 1):
        if n % i == 0:
            if i == n // i:
                factors.append(i)
            else:
                factors.append(i)
                factors.append(n // i)
    return factors

def divide_out_factors(n, factors):
    for factor in factors:
        if factor != 1 and n % (factor ** 2) == 0:
            n = n // (factor ** 2)
    return n

def factor_square_root(n):
    root = int(math.sqrt(n))
    if root * root == n:
        return root
    else:
        factors = square_factors(n)
        perfect_square = 1
        for factor in factors:
            if factor != 1:
                perfect_square *= factor
        return perfect_square, n // (perfect_square ** 2)

def rounding(num, digits):
    return round(num, digits)

def convert_to_fraction(decimal):
    return Fraction(decimal).limit_denominator()

def convert_to_percent(decimal):
    return decimal * 100

def define_function(expression):
    exec(expression)

def function_with_input(expression, x):
    return eval(expression)

def function_with_two_inputs(expression, x, y):
    return eval(expression)

def function_with_return_value(a, b):
    return a + b

def solve_for_x(expression):
    return eval(expression)

def make_your_own_function():
    expression = input("Enter your function: ")
    return lambda x: eval(expression)

def create_menu():
    while True:
        print("1. Add")
        print("2. Subtract")
        print("3. Multiply")
        print("4. Divide")
        print("5. Cast Input")
        print("6. Cast Input on the Same Line")
        print("7. Orders of Operations")
        print("8. Modulus and Remainder")
        print("9. Modulus and Factors")
        print("10. Check Prime Number")
        print("11. Find Factors")
        print("12. Reciprocals")
        print("13. Square Number")
        print("14. Square Root")
        print("15. Floor Function")
        print("16. Square Factors")
        print("17. Divide Out Factors")
        print("18. Factor Square Root")
        print("19. Rounding")
        print("20. Convert to Fraction")
        print("21. Convert to Percent")
        print("22. Define a Function")
        print("23. Function with Input")
        print("24. Function with Two Inputs")
        print("25. Function with Return Value")
        print("26. Solve for x")
        print("27. Make Your Own Function")
        print("28. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            num1, num2 = cast_input_on_same_line("Enter two numbers separated by space: ")
            result = add(num1, num2)
            print(f"The sum is: {result}")

        elif choice == '2':
            num1, num2 = cast_input_on_same_line("Enter two numbers separated by space: ")
            result = subtract(num1, num2)
            print(f"The difference is: {result}")

        elif choice == '3':
            num1, num2 = cast_input_on_same_line("Enter two numbers separated by space: ")
            result = multiply(num1, num2)
            print(f"The product is: {result}")

        elif choice == '4':
            num1, num2 = cast_input_on_same_line("Enter two numbers separated by space: ")
            result = divide(num1, num2)
            print(f"The quotient is: {result}")

        elif choice == '5':
            prompt = input("Enter a number: ")
            result = cast_input(prompt)
            print(f"The input number is: {result}")

        elif choice == '6':
            prompt = "Enter two numbers separated by space: "
            num1, num2 = cast_input_on_same_line(prompt)
            print(f"The input numbers are: {num1} and {num2}")

        elif choice == '7':
            expression = input("Enter the expression: ")
            result = orders_of_operations(expression)
            print(f"The result is: {result}")

        elif choice == '8':
            num1, num2 = cast_input_on_same_line("Enter two numbers separated by space: ")
            result = modulus_and_remainder(num1, num2)
            print(f"The remainder and quotient are: {result}")

        elif choice == '9':
            n = int(input("Enter a number: "))
            factors = modulus_and_factors(n)
            print(f"The factors of {n} are: {factors}")

        elif choice == '10':
            n = int(input("Enter a number: "))
            if is_prime(n):
                print(f"{n} is a prime number.")
            else:
                print(f"{n} is not a prime number.")

        elif choice == '11':
            n = int(input("Enter a number: "))
            factors = find_factors(n)
            print(f"The factors of {n} are: {factors}")

        elif choice == '12':
            num1, num2 = cast_input_on_same_line("Enter two numbers separated by space: ")
            result = reciprocals(num1, num2)
            print(f"The reciprocals are: {result}")

        elif choice == '13':
            num = cast_input("Enter a number: ")
            result = square(num)
            print(f"The square is: {result}")

        elif choice == '14':
            num = cast_input("Enter a number: ")
            result = square_root(num)
            print(f"The square root is: {result}")

        elif choice == '15':
            num = cast_input("Enter a number: ")
            result = floor(num)
            print(f"The floor value is: {result}")

        elif choice == '16':
            n = int(input("Enter a number: "))
            factors = square_factors(n)
            print(f"The square factors of {n} are: {factors}")

        elif choice == '17':
            n = int(input("Enter a number: "))
            factors = square_factors(n)
            result = divide_out_factors(n, factors)
            print(f"The result after dividing out factors is: {result}")

        elif choice == '18':
            n = int(input("Enter a number: "))
            result = factor_square_root(n)
            print(f"The factors of the square root of {n} are: {result}")

        elif choice == '19':
            num = cast_input("Enter a number: ")
            digits = int(input("Enter the number of digits to round to: "))
            result = rounding(num, digits)
            print(f"The rounded value is: {result}")

        elif choice == '20':
            decimal = cast_input("Enter a decimal number: ")
            result = convert_to_fraction(decimal)
            print(f"The fraction is: {result}")

        elif choice == '21':
            decimal = cast_input("Enter a decimal number: ")
            result = convert_to_percent(decimal)
            print(f"The percent is: {result}%")

        elif choice == '22':
            expression = input("Enter the function definition: ")
            define_function(expression)
            print("Function defined successfully!")

        elif choice == '23':
            expression = input("Enter the function with input 'x': ")
            x = float(input("Enter the value of 'x': "))
            result = function_with_input(expression, x)
            print(f"The result is: {result}")

        elif choice == '24':
            expression = input("Enter the function with inputs 'x' and 'y': ")
            x = float(input("Enter the value of 'x': "))
            y = float(input("Enter the value of 'y': "))
            result = function_with_two_inputs(expression, x, y)
            print(f"The result is: {result}")

        elif choice == '25':
            num1, num2 = cast_input_on_same_line("Enter two numbers separated by space: ")
            result = function_with_return_value(num1, num2)
            print(f"The result is: {result}")

        elif choice == '26':
            expression = input("Enter the equation to solve for x: ")
            result = solve_for_x(expression)
            print(f"The value of x is: {result}")

        elif choice == '27':
            custom_function = make_your_own_function()
            x = float(input("Enter a value for x: "))
            result = custom_function(x)
            print(f"The result is: {result}")

        elif choice == '28':
            print("Exiting...")
            break

        else:
            print("Invalid input. Please try again.")

if __name__ == "__main__":
    create_menu()
```
